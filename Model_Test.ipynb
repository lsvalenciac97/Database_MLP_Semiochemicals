##Load the model

```python
#Import the model from the directory in which it is stored
PathModel = '/content/drive/MyDrive/Maestría-Laura/Articulos/Supplementary_material/model_60_20_40.sav'
mlp_model = pickle.load(open(PathModel, 'rb'))
```

## Prediction - Classification of insect/molecule relationships


```python
SemiochemicalType=mlp_model.predict(X)
```

```python
Relationships = TestData_Clean[['Species','Molecule']]
```

```python
Classification=[]
for i in range(len(X)):
	Classification.append(SemiochemicalType[i])
```

```python
Relationships['Classification']=Classification
```

```python
#Correspondence of each class represented by a numerical value to the type of semiochemical
SemiochemicalType_Correspondence = {1: 'Kairomone', 6:'Sexual pheromone',
                                    7:'Aggregation pheromone', 8:'Trail pheromone',
                                    14: 'Repellent allomone', 18: 'Contact sex pheromone', 19: 'Defense allomone'}

Relationships= Relationships.replace({'Classification':SemiochemicalType_Correspondence})
```

```python
#Calculate the probabilities that an insect molecule relationship belongs to each class of semiochemical.
class_probabilities = mlp_model.predict_proba(X)
```

```python
probabilities_by_class= pd.DataFrame(class_probabilities, columns=['Kairomone probability', 'Sexual pheromone probability',
                                           'Aggregation pheromone probability', 'Trail pheromone probability',
                                           'Repellent allomone probability',
                                           'Contact sex pheromone probability', 'Defense allomone probability' ])
```

```python
pd.concat([Relationships, probabilities_by_class], axis=1)
```
